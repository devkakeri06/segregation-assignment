```python
import pandas as pd

df = pd.read_excel("/content/Assignment.xlsx")
df.head()
```

```python
df['id'] = df.groupby(['Name']).ngroup()
df = df[['id', 'Name', 'Email ID', 'Phone Number ( Whatsapp )', 'Domain']]
df.head(10)
```

```python
AIML = df[df['Domain'].str.contains('AI/ ML')]
AIML.head(10)
```

```python
!pip install xlsxwriter

CLOUD = df[df['Domain'].str.contains('Cloud')]
WEBDEV = df[df['Domain'].str.contains('Web Development')]
UIUX = df[df['Domain'].str.contains('UI/UX')]
CP = df[df['Domain'].str.contains('Competitive Programming')]
BLOCKCHAIN = df[df['Domain'].str.contains('Blockchain')]
ANDROID = df[df['Domain'].str.contains('Android')]
CREATIVE_DESIGN = df[df['Domain'].str.contains('Creative Design')]
VIDEOEDIT = df[df['Domain'].str.contains('Video Editing')]
CONTENT = df[df['Domain'].str.contains('Content Writing')]
PRA = df[df['Domain'].str.contains('PRA')]
```

```python
with pd.ExcelWriter('Devaansh_Kakeri_Assignment.xlsx') as writer:  # doctest: +SKIP
     df.to_excel(writer, sheet_name='Ass')
     AIML.to_excel(writer, sheet_name='AIML', index=False)
     CLOUD.to_excel(writer, sheet_name='CLOUD', index=False)
     WEBDEV.to_excel(writer, sheet_name='WEBDEV', index=False)
     UIUX.to_excel(writer, sheet_name='UIUX', index=False)
     CP.to_excel(writer, sheet_name='CP', index=False)
     BLOCKCHAIN.to_excel(writer, sheet_name='BLOCKCHAIN', index=False)
     CREATIVE_DESIGN.to_excel(writer, sheet_name='CREATIVE_DESIGN', index=False)
     VIDEOEDIT.to_excel(writer, sheet_name='VIDEOEDIT', index=False)
     CONTENT.to_excel(writer, sheet_name='CONTENT', index=False)
     PRA.to_excel(writer, sheet_name='PRA', index=False)

```
